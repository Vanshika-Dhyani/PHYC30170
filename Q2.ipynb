```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def signal_func(A,v,t,p,n_pts):
    y = []
    t = np.linspace(t[0],t[1],num=n_pts)
    return A*np.cos(2*np.pi*v*t + p),t
```

# 2a

```python
y,t = signal_func(1,10,[0,3],30,500)
t.shape
```

```python
%matplotlib inline
plt.figure()
plt.plot(t,y)
plt.show()
```

# 2b

```python
y1,t1 = signal_func(1,10,[0,3],30,500)
y2,t2 = signal_func(0.8,8,[0,3],60,500)
```

```python
plt.figure(figsize=(10,5))
plt.plot(t1,y1,label='A:1,v:10Hz,phase:30')
plt.plot(t2,y2,label='A:0.8,v:8Hz,phase:60')
plt.legend()
plt.show()
```

# 2c

```python
result_wave = y1+y2
```

```python
plt.figure()
plt.plot(t1,result_wave)
plt.show()
```

```python
plt.figure()
plt.plot(t1,y1,label='A:1,v:10Hz,phase:30')
plt.plot(t2,y2,label='A:0.8,v:8Hz,phase:60')
plt.plot(t1,result_wave,label='Result wave')
plt.legend()
plt.show()
```

# 2d

```python
def fft(signal_y):
    fft_y = np.fft.fft(signal_y)
    fft_y = np.abs(fft_y[:signal_y.shape[0]//2])
    
    return fft_y
```

```python
fft_y1 = fft(y1)
```

```python
fft_y2 = fft(y2)
```

```python
plt.figure()
plt.plot(fft_y1,label='FFT of y1 signal')
plt.plot(fft_y2,label='FFT of y2 signal')
plt.legend()
plt.show()
```
