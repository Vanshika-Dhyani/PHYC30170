```
import numpy as np
import math
import matplotlib.pyplot as plt
```

```
def estimate_pi(N):
  x = np.random.rand(N)
  y = np.random.rand(N)
  
  plot(x,y)

  n = 0
  for i,j in zip(x,y):
    if math.sqrt(i**2 + j**2) <= 1:
      n += 1
      
  pi = (n/N) * 4
  return pi
```

```
def plot(x,y):
  figure, axes = plt.subplots()
  draw_circle = plt.Circle((0, 0), 1, fill=False, color='r')

  axes.set_aspect(1)
  axes.add_artist(draw_circle)

  plt.scatter(x,y, s=0.5)
  plt.show()
```

```
estimate_pi(10000)
```

```
import random
import math

def inCircle(x , y):
    if  math.sqrt((x-0.5)**2 + (y-0.5)**2) <= 0.5:
        return True;
    else:
        return False;


cntInCircle = 0;
n = 10000000;

for i in range(n):
    x = random.random();
    y = random.random();
    
    if inCircle(x,y):
        cntInCircle += 1;
        
print (cntInCircle/n * 4); # M'/=pi/4
```
